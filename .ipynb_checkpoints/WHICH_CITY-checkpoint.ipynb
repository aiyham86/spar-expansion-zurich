```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Load the full dataset
df = pd.read_csv('all_supermarkets_with_size_and_rating.csv')

# Focus only on Zürich district
df_zurich = df[df['district'] == 'Zürich'].copy()

# Map size categories to numeric scores
size_map = {'Small': 1, 'Medium': 2, 'Large': 3}
df_zurich['size_score'] = df_zurich['size_category'].map(size_map)

# Group by city
city_summary = df_zurich.groupby('city').agg(
    total_stores=('name', 'count'),
    spar_count=('type', lambda x: sum(x.str.lower() == 'spar')),
    competitor_count=('type', lambda x: sum(x.str.lower() != 'spar')),
    total_size_score=('size_score', 'sum')
).reset_index()

# Calculate people per spar unit (placeholder if you had population data)
# city_summary['people_per_spar_unit'] = city_summary['population'] / (city_summary['spar_count'] + 1)

# Normalize and score
scaler = MinMaxScaler()
city_summary[['size_scaled', 'comp_scaled']] = scaler.fit_transform(
    city_summary[['total_size_score', 'competitor_count']]
)

# Final score: lower size & fewer competitors + no SPAR = better
city_summary['score'] = (
    (1 - city_summary['size_scaled']) * 0.5 +
    (1 - city_summary['comp_scaled']) * 0.3 +
    (1 - (city_summary['spar_count'] / (city_summary['total_stores'] + 1))) * 0.2
)

# Sort and show top 3 cities
top_cities = city_summary.sort_values(by='score', ascending=False).head(3)
print(top_cities[['city', 'score', 'spar_count', 'competitor_count', 'total_size_score']])
```

**Generate Grid & Calculate Distances**

```python
import pandas as pd
import numpy as np
from geopy.distance import geodesic

# Load your cleaned Zurich supermarket data
df = pd.read_csv('all_supermarkets_with_size_and_rating.csv')
df_zurich_city = df[(df['district'] == 'Zürich') & (df['city'] == 'Zürich')].copy()

# Filter SPAR and competitor stores
df_spar = df_zurich_city[df_zurich_city['type'].str.lower() == 'spar']
df_competitors = df_zurich_city[df_zurich_city['type'].str.lower().isin(['migros', 'coop', 'lidl', 'aldi', 'denner'])]

# Define Zurich city bounding box
lat_min, lat_max = 47.33, 47.43
lon_min, lon_max = 8.47, 8.62
lat_step = 0.005
lon_step = 0.005

# Generate grid points
lat_points = np.arange(lat_min, lat_max, lat_step)
lon_points = np.arange(lon_min, lon_max, lon_step)
grid_points = [(lat, lon) for lat in lat_points for lon in lon_points]
df_grid = pd.DataFrame(grid_points, columns=['lat', 'lng'])

# Functions to compute distances and counts
def min_distance(point, stores):
    return min(geodesic(point, (row['lat'], row['lng'])).meters for _, row in stores.iterrows()) if not stores.empty else None

def count_within_radius(point, stores, radius=1000):
    return sum(geodesic(point, (row['lat'], row['lng'])).meters <= radius for _, row in stores.iterrows())

# Apply to grid points
df_grid['dist_to_nearest_spar'] = df_grid.apply(lambda row: min_distance((row['lat'], row['lng']), df_spar), axis=1)
df_grid['dist_to_nearest_competitor'] = df_grid.apply(lambda row: min_distance((row['lat'], row['lng']), df_competitors), axis=1)
df_grid['spar_within_1km'] = df_grid.apply(lambda row: count_within_radius((row['lat'], row['lng']), df_spar), axis=1)
df_grid['competitors_within_1km'] = df_grid.apply(lambda row: count_within_radius((row['lat'], row['lng']), df_competitors), axis=1)

# Preview result
df_grid.head()
```

let’s now calculate a final score for each candidate location in df_grid based on our strategic goals.

| Metric                         | Direction        | Weight | Rationale                        |
| ------------------------------ | ---------------- | ------ | -------------------------------- |
| Distance to nearest SPAR       | Higher is better | 0.4    | Avoid cannibalizing nearby SPARs |
| SPAR stores within 1 km        | Lower is better  | 0.2    | Underserved area                 |
| Competitors within 1 km        | Moderate is good | 0.3    | Demand proven but not saturated  |
| Distance to nearest competitor | Lower is worse   | 0.1    | Too close = high competition     |


```python
from sklearn.preprocessing import MinMaxScaler

# Copy the grid to avoid changing the original
df_score = df_grid.copy()

# Normalize numeric values
scaler = MinMaxScaler()

# Normalize distance columns
df_score[['spar_dist_scaled', 'comp_dist_scaled']] = scaler.fit_transform(
    df_score[['dist_to_nearest_spar', 'dist_to_nearest_competitor']]
)

# Invert competitor distance (closer competitor = worse score)
df_score['comp_dist_scaled'] = 1 - df_score['comp_dist_scaled']

# Normalize count columns
df_score[['spar_near_scaled', 'comp_near_scaled']] = scaler.fit_transform(
    df_score[['spar_within_1km', 'competitors_within_1km']]
)

# Define weights
w1 = 0.4  # distance to nearest SPAR (higher is better)
w2 = 0.2  # fewer SPAR nearby
w3 = 0.3  # some competition = good
w4 = 0.1  # closer competitors = slightly worse

# Compute final score
df_score['score'] = (
    w1 * df_score['spar_dist_scaled'] -
    w2 * df_score['spar_near_scaled'] +
    w3 * df_score['comp_near_scaled'] -
    w4 * df_score['comp_dist_scaled']
)

# Sort and show top 10 locations
top_locations = df_score.sort_values(by='score', ascending=False).head(10)
print(top_locations[['lat', 'lng', 'score', 'dist_to_nearest_spar', 'spar_within_1km', 'competitors_within_1km']])
```

**Distance to Nearest Train Station**

```python
import geopandas as gpd
from geopy.distance import geodesic

# Load train station GeoJSON
gdf_stations = gpd.read_file('train_st.geojson')

# Extract coordinates
station_coords = list(zip(gdf_stations.geometry.y, gdf_stations.geometry.x))  # (lat, lng)

# Function to compute distance to nearest station
def distance_to_station(point, station_list):
    return min(geodesic(point, s).meters for s in station_list)

# Apply to grid
df_grid['dist_to_station'] = df_grid.apply(
    lambda row: distance_to_station((row['lat'], row['lng']), station_coords), axis=1
)

# Preview
df_grid[['lat', 'lng', 'dist_to_station']].head()
```

**Station Proximity to Scoring Formula**

We’ll treat closer distance to train station as better, so we’ll:  
Normalize dist_to_station  
Invert it: closer station = higher score  
Add it to your existing score with a strong weight (since you said it's most important)


```python
from sklearn.preprocessing import MinMaxScaler
from shapely.geometry import Point
import geopandas as gpd

# Create a fresh copy to avoid overwriting original
df_score = df_grid.copy()

# Normalize distance and count features
scaler = MinMaxScaler()
df_score[['spar_dist_scaled', 'comp_dist_scaled', 'station_dist_scaled']] = scaler.fit_transform(
    df_score[['dist_to_nearest_spar', 'dist_to_nearest_competitor', 'dist_to_station']]
)

# Invert competitor distance and station distance
df_score['comp_dist_scaled'] = 1 - df_score['comp_dist_scaled']  # closer = worse
df_score['station_score'] = 1 - df_score['station_dist_scaled']  # closer = better

# Normalize counts
df_score[['spar_near_scaled', 'comp_near_scaled']] = scaler.fit_transform(
    df_score[['spar_within_1km', 'competitors_within_1km']]
)

# Set weights
w_spar_dist = 0.3
w_spar_near = 0.2
w_comp_near = 0.2
w_comp_dist = 0.1
w_station = 0.4  # most important

# Compute final score
df_score['score'] = (
    w_spar_dist * df_score['spar_dist_scaled'] -
    w_spar_near * df_score['spar_near_scaled'] +
    w_comp_near * df_score['comp_near_scaled'] -
    w_comp_dist * df_score['comp_dist_scaled'] +
    w_station * df_score['station_score']
)

# Show top 10 ranked locations
top_locations = df_score.sort_values(by='score', ascending=False).head(10)


print(top_locations[['lat', 'lng', 'score', 'dist_to_station', 'dist_to_nearest_spar', 'spar_within_1km']])
```
